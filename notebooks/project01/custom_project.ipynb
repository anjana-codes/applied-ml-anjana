```python
# Display target variable histogram
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load dataset from your path
df = pd.read_csv("c:/repos/applied-ml-anjana/data/Housing.csv")
```

```python
# Characterize and plot histogram for 'price'
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="price", bins=30, kde=True)
plt.title("Histogram of Housing Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.axvline(df["price"].mean(), color="red", linestyle="--", label="Mean")
plt.legend()
plt.show()
```

```python
# Set style for consistency
sns.set_style("whitegrid")

# Plot histograms for x features: same style (all with KDE, consistent bins/params), different colors
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Area (blue, KDE=True for continuous)
sns.histplot(data=df, x="area", bins=20, kde=True, color="blue", ax=axes[0])
axes[0].set_title("Area Histogram")
axes[0].set_xlabel("Area (sq ft)")

# Bedrooms (green, KDE=True even for discrete)
sns.histplot(data=df, x="bedrooms", bins=6, kde=True, color="green", ax=axes[1])
axes[1].set_title("Bedrooms Histogram")
axes[1].set_xlabel("Number of Bedrooms")

# Bathrooms (orange, KDE=True for consistency)
sns.histplot(data=df, x="bathrooms", bins=4, kde=True, color="orange", ax=axes[2])
axes[2].set_title("Bathrooms Histogram")
axes[2].set_xlabel("Number of Bathrooms")

plt.tight_layout()
plt.savefig("x_histograms_uniform_style_different_colors.png")  # Saves as screenshot file
plt.show()

# Print stats for reference
print(df[["area", "bedrooms", "bathrooms"]].describe())
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Apply log transformation to 'area' (assuming no zeros; all positive)
df["log_area"] = np.log(df["area"])

# Plot side-by-side: Original vs. Log-transformed histograms
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Original area
sns.histplot(data=df, x="area", bins=20, kde=True, color="blue", ax=axes[0])
axes[0].set_title("Original Area Histogram")
axes[0].set_xlabel("Area (sq ft)")

# Log-transformed area
sns.histplot(data=df, x="log_area", bins=20, kde=True, color="purple", ax=axes[1])
axes[1].set_title("Log(Area) Histogram")
axes[1].set_xlabel("Log(Area)")

plt.tight_layout()
plt.savefig("area_log_transformation.png")  # Saves as screenshot file
plt.show()

# Print stats for comparison
print("Original Area Stats:")
print(df["area"].describe())
print("\nLog(Area) Stats:")
print(df["log_area"].describe())
```
